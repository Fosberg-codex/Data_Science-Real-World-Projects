## Project Suplimentary File: Some visualizations of Weather AUS Dataset

### An over view of data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

weather_dataframe = pd.read_csv('weatherAUS.csv')
```

```python
weather_dataframe.head()
```

```python
correlation = weather_dataframe.corr()
plt.figure(figsize=(16,12))
# plt.title('Figure 2.1.a Correlation Heatmap of Rain in Australia Dataset', fontsize='14')
ax = sns.heatmap(correlation, square=True, annot=True, fmt='.2f', linecolor='white', annot_kws={"size": 12})
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, fontsize='14')
ax.set_yticklabels(ax.get_yticklabels(), rotation=30, fontsize='14')           
plt.show()
```

### Some violin plots

```python
sns.violinplot(x="RainTomorrow", y="Humidity3pm", data=weather_dataframe, palette="Pastel1", annot=True)
```

```python
sns.violinplot(x="MinTemp", y="RainTomorrow", data=weather_dataframe, palette="Pastel1", annot=True)
```

```python
sns.violinplot(x="Rainfall", y="RainTomorrow", data=weather_dataframe, palette="Pastel1", annot=True)
```

### Bar chart

Just like in Bar Chart in KNIME, we can count number of Y and N using following plot:

```python
plt.figure(figsize=(6,10))
sns.countplot(data=weather_dataframe,x='RainTomorrow', palette=["#7cafb1", "#2c3e66"])
```
